```python
import csv
```

```python
with open('HTMLPull.csv', 'r') as inCSV:
    reader = csv.DictReader(inCSV,delimiter=',')
    reader = list(reader)
    for row in  reader:
        row['EnterName'] = row['EnterName'].split('>')[1].strip(' ')
        row['ExitName'] = row['ExitName'].split('>')[1].strip(' ')
        row['Distance/Km'] = row['Distance/Km'].split('>')[1].strip(' km')
        row['NormalPrice'] = row['NormalPrice'].split('>')[1].strip('$')
        row['NormalPrice'] = int(float(row['NormalPrice'])*100)
        row['EarlyPrice'] = row['NormalPrice']-50
        if row['EnterName'] == row['ExitName']:
            row['Distance/Km'] = '0'
            row['NormalPrice'] = 0
            row['EarlyPrice'] = 0
```

```python
newDict = {}
with open('cleaned.csv', 'r') as inCSV:
    reader = csv.DictReader(inCSV,delimiter=',')
    reader = list(reader)
    for row in  reader:
        if row['EnterCode']==row['ExitCode']:
            newDict[row['EnterCode'],row['ExitCode']] = row
            continue
        else:
            newDict[row['EnterCode'],row['ExitCode']] = row
            newrow = row.copy()
            newrow['EnterCode'],newrow['ExitCode'] = row['ExitCode'],row['EnterCode']
            newrow['EnterName'],newrow['ExitName'] = row['ExitName'],row['EnterName']
            newrow['EnterSign'],newrow['ExitSign'] = row['ExitSign'],row['EnterSign']
            newDict[newrow['EnterCode'],newrow['ExitCode']] = newrow
```

```python
fieldSeq = [
    'EnterCode','EnterName','EnterSign','ExitCode','ExitName','ExitSign','Distance/Km','NormalPrice','EarlyPrice'
           ]
expList = list(newDict.values())
expList = [[i[j].strip() for j in fieldSeq]for i in expList]
pd.DataFrame(expList,columns=fieldSeq).sort_values(by=['EnterName','ExitName']).to_csv('AllCleaned.csv',index=False)
```
