```python
#module imports
import geopandas as gp
import pandas as pd
import matplotlib.pyplot as plt
import geoplot
# SG map from
# https://mapcruzin.com/free-singapore-country-city-place-gis-shapefiles.htm
```

```python
'''
level 3 : 

Early Days / uncertain timings
'''
```

```python
#data ingest
Stops = gp.read_file(
    './TrainStation_Oct2017/MRTLRTStnPtt.shp'
)

Background = gp.read_file(
    './singapore_coastline/singapore_coastline.shp'
)
Background2 = gp.read_file(
    './national-map-polygon/SG_FROM_KML_BUFF.shp'
)

Price = pd.read_csv(
    'AllCleaned.csv'
)

print(Stops.crs)
print(Background.crs)
#Background = Background.to_crs(Stops.crs)
#Background2 = Background2.to_crs(Stops.crs)
Stops = Stops.to_crs(Background2.crs)
```

```python
#Filter To current only get Target
TargetStation = 'CC24'
ThisPrice = Price.loc[Price['ExitSign'] == TargetStation]
```

```python
#merge
ThisPrice['STN_NAME'] = ThisPrice['EnterName'].str.upper()
Stops['STN_NAME'] = Stops['STN_NAME'].str.split(' MRT').str[0].str.split(' LRT').str[0]

newdf = pd.merge(
    left = Stops,
    right = ThisPrice,
    on = 'STN_NAME',
    how='right'
)
newdf = newdf.drop_duplicates(subset=['STN_NAME'],keep='first')
```

```python
for i in [16,20,24,12]:
    def checkWorth(row):
        times = i
        #print(times)
        return row.NormalPrice*times*2>4500
    
    newdf['WorthIt'] = newdf.apply(lambda row: checkWorth(row), axis=1)
    newdf.to_csv('V_' + str(i) + ' days of travel a month.csv')
    '''
    fig, ax = plt.subplots(figsize=(20,20))
    geoplot.voronoi(
        newdf,
        ax=ax,
        clip=Background2,
        hue='WorthIt',
        cmap='RdYlGn',
        extent = tuple(Background2.total_bounds),
        #figsize=(20,20)
    )
    
    ax.set_title(str(i)+' days of travel a month')
    #newdf.plot(column='assess',cmap='RdYlGn',ax=ax)
    Background2.plot(ax=ax)
    #plt.show()
    fig.savefig('V_' + str(i)+' days of travel a month')
    '''
```

```python
test = newdf.buffer(0.025)
test = gp.GeoDataFrame(geometry=test)
```

```python
test2 = gp.overlay(test, Background2, how='intersection')
```

```python
fig2, ax2 = plt.subplots(figsize=(20,20))
test2.plot(ax=ax2)
geoplot.voronoi(
        newdf,
        ax=ax2,
        clip=test2,
        #hue='assess',
        #cmap='RdYlGn',
        extent = tuple(Background2.total_bounds),
        color='blue'
        #figsize=(20,20)
    )
Background2.plot(ax=ax2,color='red')
fig2.show()
```

```python
for i in [16,20,24]:
    def checkWorth(row):
        times = i
        #print(row.NormalPrice*times)
        return row.NormalPrice*times*2>4500

    fig, ax = plt.subplots(figsize=(20,20))
    ax.set_title(str(i)+' days of travel a month')
    newdf['assess'] = newdf.apply(lambda row: checkWorth(row), axis=1)
    newdf.plot(column='assess',cmap='RdYlGn',ax=ax)
    Background.plot(ax=ax)
    fig.savefig(str(i)+' days of travel a month')
```
