```python
import requests
import sys
```

```python
def get_codes(entryCode,exitCode):
    #sampleURL = "https://www.mytransport.sg/content/mytransport/home/commuting/busservices/jcr:content/par/mtp_generic_tab/mtp_generic_tab6/distancefare.mrtFare?entryMarkerId=7_7&exitMarkerId=5_5&ticketType=30"
    toKeep = [
    "distance_fare_result_bd",
    "distance_fare_result_al",
    "distance_fare_result_dt",
    "distance_fare_result_fr"
    ]
    r = requests.get(
        """https://www.mytransport.sg/content/mytransport/home/commuting/busservices/jcr:content/par/mtp_generic_tab/mtp_generic_tab6/distancefare.mrtFare?"""
        ,params={
        'entryMarkerId':str(entryCode)+'_'+str(entryCode),
        'exitMarkerId':str(exitCode)+'_'+str(exitCode),
        'ticketType':'30',
        })
    a = str(r.content)
    outPut = [entryCode,exitCode]
    for i in a.split("<"):
        for j in toKeep:
            if j in i:
                outPut.append(i)    
    return(outPut)

```

```python
newRows = []
with open('scrapee.txt') as inFile:
    for row in inFile:
        newRow = (row.split('>'))
        #del newRow[2]
        newRow[0] = newRow[0].split('"')[1]
        newRow[1] = newRow[1].split('<')[0]
        newRow[1],newRow[2] = newRow[1].split('(')[0].strip(' '),'('+ newRow[1].split('(')[1].strip(' ')
        newRow = tuple(newRow)
        if newRow not in newRows:
            newRows.append(newRow)
```

```python
stationCodes = [i[0] for i in newRows]
print (stationCodes)
```

```python
newScrape2 = []
#for i in range(5):
#    for j in range(5):
for i in range(len(stationCodes)):
    for j in range(i,len(stationCodes)):
        newScrape2.append(get_codes(stationCodes[i],stationCodes[j]))
    print(i)
```
